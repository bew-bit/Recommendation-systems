**А точно нужно сортировать по вероятности?**<br/>
Можно сортировать по вероятности, а можно отсортировать товары по маржинальности и рассылать клиентам товары с максимальной маржой: при одинаковых количественных продажах рекомендованных товаров наиболее маржинальные принесут большую прибыль.
Однако цель проекта - повысить выручку, поэтому для начала можно остановиться на сортировке по вероятности.<br/><br/>
**Какую метрику использовать?**<br/>
1. Для оценки модели можно использоветь precision@k в качестве основной и критерия прриемки заказчиком/бизнесом. Precision показывает долю правильных прогнозов из всех прогнозов. В нашем случае k=5, поэтому при высоком precision потребитель/юзер будет получать короткий список очень релевантных ему рекомендаций.<br/>
2. Очень важно следить за долей прочитанных email'ов для выбора лучшегго времени рассылки.<br/>
3. Т.к. цель проекта - увеличить выручку, так же следуетсчитать процент роста увеличения выручки по отношению к тому же периоду в прошлом году (чтобы исключить влияние сезонности).<br/><br/>
**Сколько раз в неделю отпрпавляем рассылку?**<br/>
Если в среднем покупка в магазине одним клиентом совершается один раз в неделю, можно на первом этапе установить интервал отправки сообщения один раз в неделю.<br/><br/>
**В какое время отправляем рассылку?**<br/>
Выбор времени рассылки зависит от психологического портрета нашего потребитея. Если наш потребитель является человеком средних лет, занятым на работе со стандартным графиком, email уведомление лучше отправлять вечером, т.к. утром и в течение дня наш потребитель будет занят делами по работе. Если наш потребитель совершает покупки раз в неделю по выходным, рассылку лучше отправлять вечером в пятницу или утром в один из выходных.<br/><br/>
**Будем отправлять одному юзеру много раз наши рекоммендации. Как добиться того, чтобы они хоть немного отличались?**<br/>
Можно отправлять рекомендации товаров, используя различные каналы для одного пользователя.<br/><br/>
**Нужно ли, чтобы в одной рассылке были *разные* товары? Как определить, что товары *разные*? Как добиться того, чтобы они были разными?**<br/>
Возможно, спасет семантическое индексирование.<br/><br/>
**Другое**<br/>
Для того, чтобы договориться о повышении выручки на 6%, нужно выяснить какой процент продаж занимают покупки, совершенные при переходу по ссылки из оповещения.
